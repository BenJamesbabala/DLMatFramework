## Scikit-image tutorial

```python
from skimage import data, io, filters
import matplotlib.pyplot as plt
```

```python
image = data.coins()
# ... or any other NumPy array!
edges = filters.sobel(image)
io.imshow(edges)
io.show()
```

```python
image = io.imread('cat.jpg')
io.imshow(image)
io.show()
print(image.shape)
```
